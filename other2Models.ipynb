```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , confusion_matrix , classification_report
```

```python
df = pd.read_csv(r'/content/BALANCED_ANEMIA_DATASET.csv')
df.head()
```

```python
y = df['Result']
x = df.drop(['Result'],axis=1)
x,y
```

```python
X_train,X_temp,Y_train,Y_temp = train_test_split(
    x,y,train_size=0.7,random_state=42, stratify = y
)
X_train,X_temp
```

```python
  X_val,X_test,Y_val,Y_test=train_test_split(
    X_temp,Y_temp,train_size=0.6,random_state=42, stratify = Y_temp
)
X_val,X_test
```

```python
model3 = KNeighborsClassifier()
model3.fit(X_train,Y_train)
```

```python
y_pred = model3.predict(X_test)
accuracy_score(Y_test,y_pred)
print(classification_report(Y_test,y_pred))
print(confusion_matrix(Y_test,y_pred))
```

```python
y_vpred = model3.predict(X_val)
accuracy_score(Y_val,y_vpred)
print(classification_report(Y_val,y_vpred))
print(confusion_matrix(Y_val,y_vpred))
```

```python
import pickle
pickle.dump(model3,open('model3.pkl','wb'))
model3 = open('model3.pkl','rb')
```

## **model4 : decision_tree**

```python
model4 = DecisionTreeClassifier()
model4.fit(X_train,Y_train)
```

```python
y2_predv = model4.predict(X_val)
accuracy_score(Y_val,y2_predv)
print(classification_report(Y_val,y2_predv))
print(confusion_matrix(Y_val,y2_predv))
```

```python
y2_predt =  model4.predict(X_test)
accuracy_score(Y_test,y2_predt)
print(classification_report(Y_test,y2_predt))
print(confusion_matrix(Y_test,y2_predt))
```

```python
pickle.dump(model4,open('model4.pkl','wb'))
model4 = open('model4.pkl','rb')
```

##**model5 :Random forest**

```python
model5 = RandomForestClassifier()
model5.fit(X_train,Y_train)
```

```python

y_predv5 = model5.predict(X_val)
accuracy_score(Y_val,y_predv5)
threshold = 0.4
y_proba = model5.predict_proba(X_val)
y_predv5 = (y_proba[:, 1] >= threshold).astype(int)
print(classification_report(Y_val,y_predv5))
print(confusion_matrix(Y_val,y_predv5))
```

```python
y_predt5 = model5.predict(X_test)
accuracy_score(Y_test,y_predt5)
print(classification_report(Y_test,y_predt5))
print(confusion_matrix(Y_test,y_predt5))
```

```python
pickle.dump(model5,open('model5.pkl','wb'))
model5 = open('model5.pkl','rb')
```
